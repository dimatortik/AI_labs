```python
import os, json
import numpy as np
import tensorflow as tf
from tensorflow.keras import layers, Model, callbacks, optimizers
from sklearn.metrics import confusion_matrix, accuracy_score, precision_score, recall_score, f1_score
import matplotlib.pyplot as plt
import seaborn as sns
import random

tf.keras.backend.clear_session()

TRAIN_DIR = "/kaggle/input/lab-5-cats-and-dogs/dataset/train"
TEST_DIR  = "/kaggle/input/lab-5-cats-and-dogs/dataset/test"

IMG_SIZE = (299, 299)
BATCH_SIZE = 16
VAL_SPLIT = 0.2
SEED = 42

EPOCHS_HEAD = 5
EPOCHS_FT = 10
LR_HEAD = 1e-3
LR_FT = 1e-5

USE_IMAGENET_WEIGHTS = True
WEIGHTS_PATH = None 

WEIGHTS_OUT = "inception_binary.weights.h5"
CLASS_INDICES_JSON = "class_indices.json"
METRICS_JSON = "metrics.json"

AUTOTUNE = tf.data.AUTOTUNE


# ---------------- DATA ----------------
def make_datasets():
    train_ds = tf.keras.utils.image_dataset_from_directory(
        TRAIN_DIR,
        validation_split=VAL_SPLIT,
        subset="training",
        seed=SEED,
        image_size=IMG_SIZE,
        batch_size=BATCH_SIZE,
        label_mode="binary",
        shuffle=True
    )

    val_ds = tf.keras.utils.image_dataset_from_directory(
        TRAIN_DIR,
        validation_split=VAL_SPLIT,
        subset="validation",
        seed=SEED,
        image_size=IMG_SIZE,
        batch_size=BATCH_SIZE,
        label_mode="binary",
        shuffle=False
    )

    test_ds = tf.keras.utils.image_dataset_from_directory(
        TEST_DIR,
        image_size=IMG_SIZE,
        batch_size=BATCH_SIZE,
        label_mode="binary",
        shuffle=False
    )

    class_names = train_ds.class_names
    class_indices = {name: i for i, name in enumerate(class_names)}
    with open(CLASS_INDICES_JSON, "w", encoding="utf-8") as f:
        json.dump(class_indices, f, ensure_ascii=False, indent=2)

    aug = tf.keras.Sequential([
        layers.RandomFlip("horizontal"),
        layers.RandomRotation(0.10),
        layers.RandomZoom(0.10),
        layers.RandomTranslation(0.05, 0.05),
        layers.RandomContrast(0.10),
    ], name="data_augmentation")

    # Preprocess for ImageNet InceptionV3 is scaling to [-1, 1]
    if USE_IMAGENET_WEIGHTS:
        preprocess = layers.Rescaling(1./127.5, offset=-1, name="preprocess_inceptionv3")
    else:
        preprocess = layers.Rescaling(1./255, name="preprocess_0_1")

    def map_train(x, y):
        x = aug(x, training=True)
        x = preprocess(x)
        return x, y

    def map_eval(x, y):
        x = preprocess(x)
        return x, y

    train_ds = train_ds.map(map_train, num_parallel_calls=AUTOTUNE).prefetch(AUTOTUNE)
    val_ds   = val_ds.map(map_eval,  num_parallel_calls=AUTOTUNE).prefetch(AUTOTUNE)
    test_ds  = test_ds.map(map_eval, num_parallel_calls=AUTOTUNE).prefetch(AUTOTUNE)

    return train_ds, val_ds, test_ds, class_names


def count_labels(ds):
    c0, c1 = 0, 0
    for _, y in ds.unbatch():
        v = int(y.numpy())
        if v == 0:
            c0 += 1
        else:
            c1 += 1
    return c0, c1


def compute_class_weight_from_ds(train_ds):
    c0, c1 = count_labels(train_ds)
    total = c0 + c1
    if c0 == 0 or c1 == 0:
        return None
    return {0: total / (2.0 * c0), 1: total / (2.0 * c1)}


# ---------------- MODEL (InceptionV3 "notop") ----------------
def conv2d_bn(x, filters, num_row, num_col,
             strides=(1, 1), padding="same", name=None):
    if name is not None:
        conv_name = name + "_conv"
        bn_name = name + "_bn"
    else:
        conv_name = None
        bn_name = None

    x = layers.Conv2D(
        filters, (num_row, num_col),
        strides=strides,
        padding=padding,
        use_bias=False,
        name=conv_name
    )(x)

    x = layers.BatchNormalization(axis=3, scale=False, name=bn_name)(x)
    x = layers.Activation("relu", name=name)(x)
    return x


def inception_v3_backbone(inp):
    # Stem
    x = conv2d_bn(inp, 32, 3, 3, strides=(2, 2), padding="valid")
    x = conv2d_bn(x, 32, 3, 3, padding="valid")
    x = conv2d_bn(x, 64, 3, 3)
    x = layers.MaxPooling2D((3, 3), strides=(2, 2))(x)

    x = conv2d_bn(x, 80, 1, 1, padding="valid")
    x = conv2d_bn(x, 192, 3, 3, padding="valid")
    x = layers.MaxPooling2D((3, 3), strides=(2, 2))(x)

    # mixed0
    b1 = conv2d_bn(x, 64, 1, 1)
    b5 = conv2d_bn(x, 48, 1, 1)
    b5 = conv2d_bn(b5, 64, 5, 5)
    b3 = conv2d_bn(x, 64, 1, 1)
    b3 = conv2d_bn(b3, 96, 3, 3)
    b3 = conv2d_bn(b3, 96, 3, 3)
    bp = layers.AveragePooling2D((3, 3), strides=(1, 1), padding="same")(x)
    bp = conv2d_bn(bp, 32, 1, 1)
    x = layers.Concatenate(axis=3, name="mixed0")([b1, b5, b3, bp])

    # mixed1
    b1 = conv2d_bn(x, 64, 1, 1)
    b5 = conv2d_bn(x, 48, 1, 1)
    b5 = conv2d_bn(b5, 64, 5, 5)
    b3 = conv2d_bn(x, 64, 1, 1)
    b3 = conv2d_bn(b3, 96, 3, 3)
    b3 = conv2d_bn(b3, 96, 3, 3)
    bp = layers.AveragePooling2D((3, 3), strides=(1, 1), padding="same")(x)
    bp = conv2d_bn(bp, 64, 1, 1)
    x = layers.Concatenate(axis=3, name="mixed1")([b1, b5, b3, bp])

    # mixed2
    b1 = conv2d_bn(x, 64, 1, 1)
    b5 = conv2d_bn(x, 48, 1, 1)
    b5 = conv2d_bn(b5, 64, 5, 5)
    b3 = conv2d_bn(x, 64, 1, 1)
    b3 = conv2d_bn(b3, 96, 3, 3)
    b3 = conv2d_bn(b3, 96, 3, 3)
    bp = layers.AveragePooling2D((3, 3), strides=(1, 1), padding="same")(x)
    bp = conv2d_bn(bp, 64, 1, 1)
    x = layers.Concatenate(axis=3, name="mixed2")([b1, b5, b3, bp])

    # mixed3 (reduction)
    b3 = conv2d_bn(x, 384, 3, 3, strides=(2, 2), padding="valid")
    b3d = conv2d_bn(x, 64, 1, 1)
    b3d = conv2d_bn(b3d, 96, 3, 3)
    b3d = conv2d_bn(b3d, 96, 3, 3, strides=(2, 2), padding="valid")
    bp = layers.MaxPooling2D((3, 3), strides=(2, 2))(x)
    x = layers.Concatenate(axis=3, name="mixed3")([b3, b3d, bp])

    # mixed4
    b1 = conv2d_bn(x, 192, 1, 1)
    b7 = conv2d_bn(x, 128, 1, 1)
    b7 = conv2d_bn(b7, 128, 1, 7)
    b7 = conv2d_bn(b7, 192, 7, 1)
    b7d = conv2d_bn(x, 128, 1, 1)
    b7d = conv2d_bn(b7d, 128, 7, 1)
    b7d = conv2d_bn(b7d, 128, 1, 7)
    b7d = conv2d_bn(b7d, 128, 7, 1)
    b7d = conv2d_bn(b7d, 192, 1, 7)
    bp = layers.AveragePooling2D((3, 3), strides=(1, 1), padding="same")(x)
    bp = conv2d_bn(bp, 192, 1, 1)
    x = layers.Concatenate(axis=3, name="mixed4")([b1, b7, b7d, bp])

    # mixed5, mixed6
    for i in range(2):
        b1 = conv2d_bn(x, 192, 1, 1)
        b7 = conv2d_bn(x, 160, 1, 1)
        b7 = conv2d_bn(b7, 160, 1, 7)
        b7 = conv2d_bn(b7, 192, 7, 1)
        b7d = conv2d_bn(x, 160, 1, 1)
        b7d = conv2d_bn(b7d, 160, 7, 1)
        b7d = conv2d_bn(b7d, 160, 1, 7)
        b7d = conv2d_bn(b7d, 160, 7, 1)
        b7d = conv2d_bn(b7d, 192, 1, 7)
        bp = layers.AveragePooling2D((3, 3), strides=(1, 1), padding="same")(x)
        bp = conv2d_bn(bp, 192, 1, 1)
        x = layers.Concatenate(axis=3, name=f"mixed{5 + i}")([b1, b7, b7d, bp])

    # mixed7
    b1 = conv2d_bn(x, 192, 1, 1)
    b7 = conv2d_bn(x, 192, 1, 1)
    b7 = conv2d_bn(b7, 192, 1, 7)
    b7 = conv2d_bn(b7, 192, 7, 1)
    b7d = conv2d_bn(x, 192, 1, 1)
    b7d = conv2d_bn(b7d, 192, 7, 1)
    b7d = conv2d_bn(b7d, 192, 1, 7)
    b7d = conv2d_bn(b7d, 192, 7, 1)
    b7d = conv2d_bn(b7d, 192, 1, 7)
    bp = layers.AveragePooling2D((3, 3), strides=(1, 1), padding="same")(x)
    bp = conv2d_bn(bp, 192, 1, 1)
    x = layers.Concatenate(axis=3, name="mixed7")([b1, b7, b7d, bp])

    # mixed8 (reduction)
    b3 = conv2d_bn(x, 192, 1, 1)
    b3 = conv2d_bn(b3, 320, 3, 3, strides=(2, 2), padding="valid")
    b7 = conv2d_bn(x, 192, 1, 1)
    b7 = conv2d_bn(b7, 192, 1, 7)
    b7 = conv2d_bn(b7, 192, 7, 1)
    b7 = conv2d_bn(b7, 192, 3, 3, strides=(2, 2), padding="valid")
    bp = layers.MaxPooling2D((3, 3), strides=(2, 2))(x)
    x = layers.Concatenate(axis=3, name="mixed8")([b3, b7, bp])

    # mixed9, mixed10
    for i in range(2):
        b1 = conv2d_bn(x, 320, 1, 1)

        b3 = conv2d_bn(x, 384, 1, 1)
        b3_1 = conv2d_bn(b3, 384, 1, 3)
        b3_2 = conv2d_bn(b3, 384, 3, 1)
        b3 = layers.Concatenate(axis=3, name=f"mixed9_{i}")([b3_1, b3_2])

        b3d = conv2d_bn(x, 448, 1, 1)
        b3d = conv2d_bn(b3d, 384, 3, 3)
        b3d_1 = conv2d_bn(b3d, 384, 1, 3)
        b3d_2 = conv2d_bn(b3d, 384, 3, 1)
        b3d = layers.Concatenate(axis=3)([b3d_1, b3d_2])

        bp = layers.AveragePooling2D((3, 3), strides=(1, 1), padding="same")(x)
        bp = conv2d_bn(bp, 192, 1, 1)

        x = layers.Concatenate(axis=3, name=f"mixed{9 + i}")([b1, b3, b3d, bp])

    return x


def build_inception_v3_binary(input_shape=(299, 299, 3), dropout=0.2):
    inp = layers.Input(shape=input_shape)
    feats = inception_v3_backbone(inp)

    x = layers.GlobalAveragePooling2D(name="gap")(feats)
    x = layers.Dropout(dropout, name="dropout")(x)
    out = layers.Dense(1, activation="sigmoid", name="predictions_binary")(x)
    model = Model(inp, out, name="inception_v3_binary")

    backbone = Model(inp, feats, name="inception_v3")
    return model, backbone


def try_load_imagenet_notop_weights(backbone):
    if not USE_IMAGENET_WEIGHTS:
        print("[INFO] Training from scratch (USE_IMAGENET_WEIGHTS=False).")
        return False

    if WEIGHTS_PATH is None:
        url = "https://storage.googleapis.com/tensorflow/keras-applications/inception_v3/inception_v3_weights_tf_dim_ordering_tf_kernels_notop.h5"
        try:
            weights_file = tf.keras.utils.get_file(
                fname=os.path.basename(url),
                origin=url,
                cache_subdir="models"
            )
        except Exception as e:
            print("[WARN] Could not download ImageNet weights:", str(e))
            print("[WARN] Set WEIGHTS_PATH to a local .h5 (e.g., Kaggle Dataset) to use pretrained weights.")
            return False
    else:
        weights_file = WEIGHTS_PATH

    try:
        backbone.load_weights(weights_file)
        print("[INFO] ImageNet 'notop' weights loaded into backbone.")
        return True
    except Exception as e:
        print("[WARN] Weight loading failed (architecture/name mismatch):", str(e))
        print("[WARN] Continue training from scratch.")
        return False


# ---------------- TRAIN / EVAL ----------------
def compile_model(model, lr):
    model.compile(
        optimizer=optimizers.Adam(learning_rate=lr),
        loss="binary_crossentropy",
        metrics=["accuracy", tf.keras.metrics.Precision(name="precision"), tf.keras.metrics.Recall(name="recall")]
    )


def eval_and_plot(model, ds, class_names, title):
    y_true = np.concatenate([y.numpy().ravel() for _, y in ds]).astype(int)
    y_prob = model.predict(ds, verbose=0).ravel()
    y_pred = (y_prob > 0.5).astype(int)

    acc = accuracy_score(y_true, y_pred)
    prec = precision_score(y_true, y_pred, zero_division=0)
    rec = recall_score(y_true, y_pred, zero_division=0)
    f1 = f1_score(y_true, y_pred, zero_division=0)

    cm = confusion_matrix(y_true, y_pred, labels=[0, 1])

    plt.figure(figsize=(5, 4))
    sns.heatmap(cm, annot=True, fmt="d", xticklabels=class_names, yticklabels=class_names)
    plt.xlabel("Predicted")
    plt.ylabel("True")
    plt.title(title)
    plt.show()

    return {"accuracy": float(acc), "precision": float(prec), "recall": float(rec), "f1": float(f1), "cm": cm.tolist()}


def predict_image(model, image_path, class_names):
    img = tf.keras.utils.load_img(image_path, target_size=IMG_SIZE)
    x = tf.keras.utils.img_to_array(img)
    x = tf.expand_dims(x, axis=0)

    if USE_IMAGENET_WEIGHTS:
        x = (x / 127.5) - 1.0
    else:
        x = x / 255.0

    prob = float(model.predict(x, verbose=0).ravel()[0])
    pred = 1 if prob > 0.5 else 0
    return {"path": image_path, "prob_class1": prob, "pred_index": pred, "pred_name": class_names[pred]}


def main():
    print("[INFO] TF version:", tf.__version__)

    train_ds, val_ds, test_ds, class_names = make_datasets()
    print("[INFO] class_names:", class_names)

    tr0, tr1 = count_labels(train_ds)
    va0, va1 = count_labels(val_ds)
    te0, te1 = count_labels(test_ds)
    print(f"[INFO] label counts train: (0={tr0}, 1={tr1})")
    print(f"[INFO] label counts val:   (0={va0}, 1={va1})")
    print(f"[INFO] label counts test:  (0={te0}, 1={te1})")

    class_weight = compute_class_weight_from_ds(train_ds)
    print("[INFO] class_weight:", class_weight)

    model, backbone = build_inception_v3_binary(input_shape=(IMG_SIZE[0], IMG_SIZE[1], 3), dropout=0.2)

    loaded = try_load_imagenet_notop_weights(backbone)
    print("[INFO] pretrained loaded:", loaded)

    print("\n[DEBUG] First 40 backbone layers:")
    for i, l in enumerate(backbone.layers[:40]):
        print(i, l.name, l.__class__.__name__, getattr(l, "kernel_size", None))

    model.summary()

    # Phase 1: train head (freeze backbone)
    for layer in model.layers:
        layer.trainable = False
    for name in ["gap", "dropout", "predictions_binary"]:
        model.get_layer(name).trainable = True

    compile_model(model, LR_HEAD)
    cb = [
        callbacks.EarlyStopping(monitor="val_loss", patience=3, restore_best_weights=True),
        callbacks.ReduceLROnPlateau(monitor="val_loss", factor=0.2, patience=2, min_lr=1e-7),
    ]
    model.fit(
        train_ds,
        validation_data=val_ds,
        epochs=EPOCHS_HEAD,
        class_weight=class_weight,
        callbacks=cb,
        verbose=1
    )

    # Phase 2: fine-tune (unfreeze last layers)
    for layer in model.layers[-80:]:
        layer.trainable = True

    compile_model(model, LR_FT)
    cb2 = [
        callbacks.EarlyStopping(monitor="val_loss", patience=5, restore_best_weights=True),
        callbacks.ReduceLROnPlateau(monitor="val_loss", factor=0.2, patience=3, min_lr=1e-7),
        callbacks.ModelCheckpoint(WEIGHTS_OUT, save_best_only=True, save_weights_only=True, monitor="val_loss"),
    ]
    model.fit(
        train_ds,
        validation_data=val_ds,
        epochs=EPOCHS_FT,
        class_weight=class_weight,
        callbacks=cb2,
        verbose=1
    )

    model.save_weights(WEIGHTS_OUT)

    print("\n[INFO] Validation metrics:")
    val_metrics = eval_and_plot(model, val_ds, class_names, "Confusion matrix (val)")

    print("\n[INFO] Test metrics:")
    test_metrics = eval_and_plot(model, test_ds, class_names, "Confusion matrix (test)")

    out = {"val": val_metrics, "test": test_metrics, "class_names": class_names}
    with open(METRICS_JSON, "w", encoding="utf-8") as f:
        json.dump(out, f, ensure_ascii=False, indent=2)

    print("\n[INFO] Saved files in /kaggle/working:")
    print(os.listdir("/kaggle/working"))

    print("\n[INFO] Single-image predictions from TEST_DIR:")

    test_files = tf.io.gfile.glob(os.path.join(TEST_DIR, "*", "*"))
    if not test_files:
        raise RuntimeError(f"No files found in TEST_DIR: {TEST_DIR}")

    for p in random.sample(test_files, k=min(10, len(test_files))):
        r = predict_image(model, p, class_names)
        print(f"{os.path.basename(p):30s} -> pred={r['pred_name']:4s}  prob_class1={r['prob_class1']:.4f}  path={r['path']}")


if __name__ == "__main__":
    main()
```
